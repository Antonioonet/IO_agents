```python
df_io_gd_twts = pd.read_csv('data/twitter-information-operations-classification/versions/1/all_tweets_hashed.csv')
```

```python
# Separate users by info_op status
io_users = df_io_gd[df_io_gd['info_op'] == 1]['retweet_ratio']
non_io_users = df_io_gd[df_io_gd['info_op'] == 0]['retweet_ratio']

# Create the plot
plt.figure(figsize=(12, 6))
plt.hist(non_io_users, bins=50, alpha=0.7, label='Non-IO Users', color='blue', edgecolor='black')
plt.hist(io_users, bins=50, alpha=0.7, label='IO Users', color='orange', edgecolor='black')

plt.xlabel('Retweet Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Retweet Ratio: IO Users vs Non-IO Users')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
features = ['avg_hashtags', 'avg_like_count', 'avg_quote_count', 'avg_urls']

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
for ax, feat in zip(axs.flat, features):
    io_vals = df_io_gd[df_io_gd['info_op'] == 1][feat]
    non_io_vals = df_io_gd[df_io_gd['info_op'] == 0][feat]

    ax.hist(non_io_vals, bins=50, alpha=0.6, label='Non-IO Users', color='blue', edgecolor='black')
    ax.hist(io_vals, bins=50, alpha=0.6, label='IO Users', color='orange', edgecolor='black')

    ax.set_title(f'Distribution of {feat.replace("_", " ").title()}')
    ax.set_xlabel(feat.replace("_", " ").title())
    ax.set_ylabel('Frequency')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
retweet_ratio_df_io = df_io['Retweeted From'].notna().mean()
comment_ratio_df_io = df_io['Commented On'].notna().mean()

like_ratio_df_io = (df_io['Likes'] > 0).mean()
hashtag_ratio_df_io = df_io['Content'].str.contains(r'#\w+', na=False).mean()

ratios = pd.Series({
    'Retweet Ratio': retweet_ratio_df_io,
    'Comment Ratio': comment_ratio_df_io,
    'Like Ratio': like_ratio_df_io,
    'Hashtag Ratio': hashtag_ratio_df_io,
})

fig, ax = plt.subplots(figsize=(8, 5))
ratios.plot(
    kind='bar',
    ax=ax,
    color=['tab:blue', 'tab:orange', 'tab:green', 'tab:purple'],
    edgecolor='black'
)
ax.set_ylim(0, 1)
ax.set_ylabel('Ratio')
ax.set_title('IO Tweet Engagement / Hashtag Ratios')
ax.grid(axis='y', alpha=0.3)

for i, v in enumerate(ratios):
    ax.text(i, v + 0.01, f'{v:.3f}', ha='center')

plt.tight_layout()
plt.show()

ratios
```
